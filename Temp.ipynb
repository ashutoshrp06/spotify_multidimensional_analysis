```python
CSV_PATH = "Spotify_Dataset_V3.csv"   
CSV_DELIMITER = ";"                   

OUT_DIR = "spotify_clustering_out"
SAMPLE_FOR_MODEL_SEL = 50000
VIS_SAMPLE = 20000
RANDOM_STATE = 42
SAVE_LABELED_CSV = True

# Clustering params
MINIBATCH_BATCH_SIZE = 10000
MINIBATCH_MAX_ITER = 200
K_AUDIO_RANGE = list(range(2,9))
K_ARTIST_RANGE = list(range(2,7))

AUDIO_FEATURES = ['Danceability','Energy','Loudness','Speechiness',
                  'Acousticness','Instrumentalness','Valence']

ARTIST_ID_COL = 'Artist (Ind.)'
ARTIST_POINTS_COL = 'Points (Ind for each Artist/Nat)'

print("Config set. CSV_PATH =", CSV_PATH)
```

```python
import os, gc, math, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics import silhouette_score
from joblib import dump

os.makedirs(OUT_DIR, exist_ok=True)

def reduce_mem_usage(df, verbose=True):
    """Downcast numeric dtypes to reduce pandas RAM usage."""
    start_mem = df.memory_usage(deep=True).sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtype
        if col_type == object:
            continue
        if pd.api.types.is_integer_dtype(col_type):
            c_min, c_max = df[col].min(), df[col].max()
            if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                df[col] = df[col].astype(np.int8)
            elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                df[col] = df[col].astype(np.int16)
            else:
                df[col] = df[col].astype(np.int32)
        else:
            df[col] = pd.to_numeric(df[col], downcast='float')
    if verbose:
        end_mem = df.memory_usage(deep=True).sum() / 1024**2
        print(f"Memory usage: {start_mem:.2f} MB -> {end_mem:.2f} MB")
    return df
```

```python
print("Reading CSV safely with semicolon delimiter...")

df = pd.read_csv(CSV_PATH, delimiter=CSV_DELIMITER, low_memory=False)
print("Initial shape:", df.shape)

df.columns = df.columns.str.strip()

numeric_candidates = AUDIO_FEATURES + [
    ARTIST_POINTS_COL, '# of Artist', '# of Nationality', 
    'Rank', 'Points (Total)'
]
for col in numeric_candidates:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

for col in numeric_candidates:
    if col in df.columns:
        print(f"{col}: {df[col].isna().sum()} NaNs after numeric conversion")

df = reduce_mem_usage(df)

print("Final dtypes:")
print(df.dtypes.head(15))
print("✅ CSV loaded safely and memory-optimized.")
```

```python
print("Basic cleaning...")
for col in ['id', 'Song URL']:
    if col in df.columns:
        df.drop(columns=[col], inplace=True)

missing_audio = [c for c in AUDIO_FEATURES if c not in df.columns]
if missing_audio:
    raise KeyError(f"Missing audio feature columns: {missing_audio}")

loud_med = df['Loudness'].median()
print("Raw Loudness median:", loud_med)
if loud_med < -1000:
    print("Converting Loudness (likely milli-DB) -> divide by 1000.")
    df['Loudness'] = df['Loudness'] / 1000.0
elif loud_med < -100:
    print("Converting Loudness (likely centi-DB) -> divide by 100.")
    df['Loudness'] = df['Loudness'] / 100.0
else:
    print("Loudness looks like decibels already.")

df['Danceability'] = df['Danceability'].clip(0,1)
df['Energy'] = df['Energy'].clip(0,1)
df['Speechiness'] = df['Speechiness'].clip(0,1)
df['Acousticness'] = df['Acousticness'].clip(0,1)
df['Instrumentalness'] = df['Instrumentalness'].clip(0,1)
df['Valence'] = df['Valence'].clip(0,1)

audio_non_na = df[AUDIO_FEATURES].notna().sum(axis=1)
df = df[audio_non_na >= 5].reset_index(drop=True)
print("Shape after requiring >=5 audio features:", df.shape)
```

```python
sample_size = min(SAMPLE_FOR_MODEL_SEL, len(df))
print("Sampling", sample_size, "rows for model selection.")
df_sample = df.sample(sample_size, random_state=RANDOM_STATE).copy()

X_sample_audio = df_sample[AUDIO_FEATURES].astype(float).values
scaler_audio = StandardScaler().fit(X_sample_audio)
X_audio_scaled_sample = scaler_audio.transform(X_sample_audio)

dump(scaler_audio, os.path.join(OUT_DIR, 'scaler_audio.joblib'))
print("Scaler fitted and saved.")
```

```python
inertias = []
sil_scores = []
ks = K_AUDIO_RANGE
print("Testing k values:", ks)

for k in ks:
    mbk = MiniBatchKMeans(n_clusters=k, batch_size=2048, random_state=RANDOM_STATE, n_init=10)
    mbk.fit(X_audio_scaled_sample)
    labels = mbk.predict(X_audio_scaled_sample)
    inertias.append(mbk.inertia_)
    try:
        sil = silhouette_score(X_audio_scaled_sample, labels, sample_size=min(5000, len(X_audio_scaled_sample)), random_state=RANDOM_STATE)
    except Exception:
        sil = float('nan')
    sil_scores.append(sil)
    print(f"k={k}: inertia={mbk.inertia_:.1f}, silhouette={sil:.4f}")

plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(ks, inertias, '-o'); plt.title("Elbow (sample)"); plt.xlabel("k"); plt.ylabel("Inertia")
plt.subplot(1,2,2); plt.plot(ks, sil_scores, '-o'); plt.title("Silhouette (sample)"); plt.xlabel("k"); plt.ylabel("Silhouette")
plt.tight_layout(); plt.show()

try:
    best_k_audio = ks[int(np.nanargmax(sil_scores))]
except Exception:
    best_k_audio = 4
print("Auto-chosen k for audio clusters:", best_k_audio)
```

```python
from sklearn.utils import shuffle
k_audio = best_k_audio
print("Fitting MiniBatchKMeans on full dataset with k_audio =", k_audio)

mbk_full = MiniBatchKMeans(n_clusters=k_audio, batch_size=MINIBATCH_BATCH_SIZE,
                           max_iter=MINIBATCH_MAX_ITER, random_state=RANDOM_STATE, n_init=10)

def iter_scaled_audio(df_obj, batch_rows=MINIBATCH_BATCH_SIZE):
    n = len(df_obj)
    for start in range(0, n, batch_rows):
        end = min(n, start + batch_rows)
        chunk = df_obj.iloc[start:end][AUDIO_FEATURES].astype(float).values
        yield scaler_audio.transform(chunk)

passes = 3  
for p in range(passes):
    print(f"Partial-fit pass {p+1}/{passes}")
    for chunk in iter_scaled_audio(df, batch_rows=MINIBATCH_BATCH_SIZE):
        mbk_full.partial_fit(chunk)

audio_labels = np.empty(len(df), dtype=np.int32)
idx = 0
for chunk in iter_scaled_audio(df, batch_rows=MINIBATCH_BATCH_SIZE):
    labels = mbk_full.predict(chunk)
    audio_labels[idx: idx + labels.shape[0]] = labels
    idx += labels.shape[0]

df['Audio_Cluster'] = audio_labels
print("Assigned audio labels. Cluster counts:")
print(df['Audio_Cluster'].value_counts())
dump(mbk_full, os.path.join(OUT_DIR, f'mbk_audio_k{k_audio}.joblib'))
print("MiniBatchKMeans saved.")
```

```python
vis_size = min(VIS_SAMPLE, len(df))
vis_df = df.sample(vis_size, random_state=RANDOM_STATE)
X_vis = scaler_audio.transform(vis_df[AUDIO_FEATURES].astype(float).values)
pca = PCA(n_components=2, random_state=RANDOM_STATE).fit(X_vis)
proj = pca.transform(X_vis)

plt.figure(figsize=(8,6))
sc = plt.scatter(proj[:,0], proj[:,1], c=vis_df['Audio_Cluster'], s=8, cmap='tab10', alpha=0.8)
plt.title("Audio clusters (PCA 2D projection of sample)")
plt.xlabel("PCA1"); plt.ylabel("PCA2")
plt.colorbar(sc, label='Audio_Cluster')
plt.tight_layout(); plt.show()
print("Explained variance ratio (PCA 2 components):", pca.explained_variance_ratio_.round(4))
```

```python

print("Aggregating artist-level metrics...")


artist_df = df.groupby(ARTIST_ID_COL).agg(
    artist_song_count = ('Title', 'count') if 'Title' in df.columns else (ARTIST_POINTS_COL, 'count'),
    avg_points = (ARTIST_POINTS_COL, 'mean'),
    total_points = (ARTIST_POINTS_COL, 'sum'),
    avg_danceability = ('Danceability', 'mean'),
    avg_energy = ('Energy', 'mean')
).reset_index()

artist_df = reduce_mem_usage(artist_df)
artist_df.head()
```

```python
artist_features = ['avg_points', 'total_points', 'artist_song_count']
X_artist = artist_df[artist_features].fillna(0).astype(float).values
scaler_artist = StandardScaler().fit(X_artist)
X_artist_scaled = scaler_artist.transform(X_artist)

inertias_a, sil_a = [], []
for k in K_ARTIST_RANGE:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10).fit(X_artist_scaled)
    inertias_a.append(km.inertia_)
    try:
        sil = silhouette_score(X_artist_scaled, km.labels_, sample_size=min(5000, len(X_artist_scaled)), random_state=RANDOM_STATE)
    except Exception:
        sil = float('nan')
    sil_a.append(sil)
    print(f"Artist k={k}: inertia={km.inertia_:.1f}, silhouette={sil:.4f}")

best_k_artist = K_ARTIST_RANGE[int(np.nanargmax(sil_a))]
print("Auto-chosen k for artists:", best_k_artist)

# Fit final KMeans on artist-level data
from joblib import dump
km_artist = KMeans(n_clusters=best_k_artist, random_state=RANDOM_STATE, n_init=20).fit(X_artist_scaled)
artist_df['Artist_Cluster'] = km_artist.labels_
dump(km_artist, os.path.join(OUT_DIR, f'kmeans_artist_k{best_k_artist}.joblib'))
dump(scaler_artist, os.path.join(OUT_DIR, 'scaler_artist.joblib'))
print("Artist KMeans model and scaler saved.")
artist_df.head()
```

```python

artist_summary = artist_df.groupby('Artist_Cluster')[artist_features].agg(['mean','median','count']).round(4)
artist_summary.to_csv(os.path.join(OUT_DIR, 'artist_cluster_summary.csv'))
print("Artist cluster summary saved:", os.path.join(OUT_DIR, 'artist_cluster_summary.csv'))
artist_summary

artist_cluster_map = artist_df.set_index(ARTIST_ID_COL)['Artist_Cluster'].to_dict()
df['Artist_Cluster'] = df[ARTIST_ID_COL].map(artist_cluster_map).fillna(-1).astype(int)
print("Merged artist cluster labels into main dataframe. Sample:")
df[[ARTIST_ID_COL, 'Artist_Cluster', 'Title']].head(10)
```

```python
print("Saving final outputs...")

cols_to_save = ['Rank','Title','Artists'] if 'Rank' in df.columns else []
cols_to_save += AUDIO_FEATURES + ['Audio_Cluster', ARTIST_ID_COL, 'Artist_Cluster', ARTIST_POINTS_COL]
cols_to_save = [c for c in cols_to_save if c in df.columns]

out_path = os.path.join(OUT_DIR, 'spotify_with_clusters_samplecols.csv')
if SAVE_LABELED_CSV:
    df[cols_to_save].to_csv(out_path, index=False)
    print("Saved labeled CSV to:", out_path)

artist_out = os.path.join(OUT_DIR, 'artist_level_with_clusters.csv')
artist_df.to_csv(artist_out, index=False)
print("Saved artist-level clusters to:", artist_out)

print("Audio and artist summaries in folder:", OUT_DIR)
```

```python
global_median = df[AUDIO_FEATURES].median()
labels = {}
for cluster, row in df.groupby('Audio_Cluster')[AUDIO_FEATURES].mean().iterrows():
    desc = []
    for feat in AUDIO_FEATURES:
        val = row[feat]
        med = global_median[feat]
        if feat in ['Danceability','Energy','Valence','Acousticness']:
            if val >= med + 0.08:
                desc.append(f"High {feat}")
            elif val <= med - 0.08:
                desc.append(f"Low {feat}")
        else:
            if val >= med + 0.05:
                desc.append(f"Higher {feat}")
            elif val <= med - 0.05:
                desc.append(f"Lower {feat}")
    labels[cluster] = ", ".join(desc) if desc else "Balanced"
print("Auto-generated audio cluster descriptions (heuristic):")
for k,v in labels.items():
    print(f"Cluster {k}: {v}")
df['Audio_Cluster_Label'] = df['Audio_Cluster'].map(labels)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertias = []
silhouettes = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(K_range, inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')

plt.subplot(1,2,2)
plt.plot(K_range, silhouettes, marker='o')
plt.title('Silhouette Scores')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Score')
plt.show()
```

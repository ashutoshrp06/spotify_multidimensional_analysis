```python
# Convert Date to datetime
ds['Date'] = pd.to_datetime(ds['Date'])

# Extract temporal features
ds['Year'] = ds['Date'].dt.year
ds['Month'] = ds['Date'].dt.month
ds['Day'] = ds['Date'].dt.day
ds['DayOfWeek'] = ds['Date'].dt.dayofweek  # Monday=0, Sunday=6
ds['DayName'] = ds['Date'].dt.day_name()
ds['WeekOfYear'] = ds['Date'].dt.isocalendar().week
ds['Quarter'] = ds['Date'].dt.quarter

# Create season feature
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

ds['Season'] = ds['Month'].apply(get_season)

# Check date range
print(f"Date Range: {ds['Date'].min()} to {ds['Date'].max()}")
print(f"Total days covered: {(ds['Date'].max() - ds['Date'].min()).days}")

# Check for gaps in dates
date_range = pd.date_range(start=ds['Date'].min(), end=ds['Date'].max(), freq='D')
missing_dates = date_range[~date_range.isin(ds['Date'])]
print(f"\nMissing dates in dataset: {len(missing_dates)}")
if len(missing_dates) > 0 and len(missing_dates) < 50:
    print("Sample missing dates:", missing_dates[:10].tolist())
```

```python
# First, check what the Artists column looks like
print("Sample Artists entries:")
print(ds['Artists'].head(20))

# Identify the delimiter (could be ',', ' & ', ' feat. ', etc.)
# Let's check if there's a pattern

print("\nMultiple artist examples:")

# Common separators: comma, ampersand, "feat", "ft"
# You'll need to adjust based on your data
def split_artists(artist_string):
    """Split artist string into list of individual artists"""
    if pd.isna(artist_string):
        return []
    
    # Replace common separators with comma
    artist_string = str(artist_string)
    separators = [' & ', ' and ', ' feat. ', ' feat ', ' ft. ', ' ft ', 
                  ' featuring ', ' with ']
    
    for sep in separators:
        artist_string = artist_string.replace(sep, ',')
    
    # Split by comma and clean
    artists = [a.strip() for a in artist_string.split(',')]
    return artists

# Apply to create new column
ds['Artists_List'] = ds['Artists'].apply(split_artists)

# Verify against '# of Artist' column
ds['Calculated_Artist_Count'] = ds['Artists_List'].apply(len)
mismatch = ds[ds['Calculated_Artist_Count'] != ds['# of Artist']]
print(f"\nMismatches between parsed and given artist count: {len(mismatch)}")

if len(mismatch) > 0:
    print("Sample mismatches (might need to adjust parsing logic):")
    print(mismatch[['Artists', '# of Artist', 'Calculated_Artist_Count']].head())

# Create a set of all unique artists
all_artists = set()
for artists_list in ds['Artists_List']:
    all_artists.update(artists_list)

print(f"\nTotal unique artists in dataset: {len(all_artists)}")
```
